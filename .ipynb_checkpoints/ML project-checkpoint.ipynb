# Import required lib

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import tree 
from sklearn.model_selection import train_test_split
from imblearn.under_sampling import RandomUnderSampler
from imblearn.over_sampling import RandomOverSampler
from sklearn.metrics import accuracy_score

from sklearn.tree import DecisionTreeClassifier

from sklearn.naive_bayes import GaussianNB

from sklearn.linear_model import LogisticRegression
```

# Read the data

```python
data = pd.read_csv('weather.csv') #read the csv file
weather_data= data.drop(columns=['date']) #drop the date column

#divide the data to input and output
weather_input=weather_data.drop(columns=['weather'])
weather_output=weather_data['weather']
```

# Preprocessing the data

```python
#this line for know how many missing value in input column#
weather_input.isnull().sum()/len(weather_input) 
```

```python
print('the number of missing value in output column: ',weather_output.isnull().sum()/len(weather_output)) 
```

```python
print('the data types of the dataset:\n',weather_data.dtypes)
print('------------------------')
print('the unique data in the output:',weather_data['weather'].unique())
```

```python
#this lines for get the weather conditions depend on the code or the value of the dictionary 
weather={'drizzle':0, 'rain':1, 'sun':2, 'snow':3, 'fog':4}
def getWeather(n):
    for key, value in weather.items():
        if n == value :
            return key
```

```python
# one-hot encoding for the remaining categorical columns: 'drizzle', 'rain', 'sun', 'snow', 'fog'
weather_output=weather_output.replace(weather)
```

# divide the data to test and validation and traning

```python
x, x_test, y, y_test = train_test_split(weather_input, weather_output, test_size=0.25, random_state=0)

x_train, x_val, y_train, y_val = train_test_split(x, y, test_size=0.30, random_state=0)
```

```python
print('X_train:', x_train.shape)
print('y_train:', y_train.shape)
print('-------------------------')
print('X_val:', x_val.shape)
print('y_val:', y_val.shape)
print('-------------------------')
print('X_test:', x_test.shape)
print('y_test:', y_test.shape)
```

# Solving the problem of imbalanced data

```python
#count the number of samples before make balance
y_train.value_counts()
```

```python
#Undersamping
rus = RandomUnderSampler(sampling_strategy={0: 28, 1:150, 2:150, 3:13, 4:30}, random_state=0)

x_train_rus, y_train_rus = rus.fit_resample(x_train, y_train)

y_train_rus.value_counts()
```

```python
#Oversamping
ros = RandomOverSampler(sampling_strategy={0: 150, 1:150, 2:150, 3:150, 4:150}, random_state=0)

x_train_balanced, y_train_balanced = ros.fit_resample(x_train_rus, y_train_rus)

y_train_balanced.value_counts()
```

# Decision Tree

```python
#test multi depth to see the best one
max_depth_values = list(range(1, 21))
acc_train_values = []
acc_val_values = []

for max_depth in max_depth_values:
    model = DecisionTreeClassifier(max_depth=max_depth, random_state=0)
    model.fit(x_train_balanced, y_train_balanced)
    y_pred_train = model.predict(x_train_balanced)
    y_pred_val = model.predict(x_val)
    acc_train = accuracy_score(y_train_balanced, y_pred_train)
    acc_val = accuracy_score(y_val, y_pred_val)
    acc_train_values.append(acc_train)
    acc_val_values.append(acc_val)
    
results = pd.DataFrame({
    'max_depth': max_depth_values,
    'acc_train': acc_train_values,
    'acc_val': acc_val_values,
})

results
```

```python
#draw the relation between max_depth and accuuracy
plt.plot(max_depth_values, acc_train_values, label='accuracy train')
plt.plot(max_depth_values, acc_val_values, label='accuracy validation ')
plt.legend()
plt.xticks(np.arange(1, 21))
plt.grid(axis='both')
plt.xlabel('max_depth')
plt.ylabel('accuracy')
plt.title('the relation between max_depth and accuuracy')
plt.show()
```

```python
#make the best decision tree model depends on the best depth that make the best accuracy 
dt_model = DecisionTreeClassifier(max_depth=3 , random_state=0)
dt_model.fit(x_train_balanced, y_train_balanced)
dt_y_pred_test = dt_model.predict(x_test)
dt_accuracy = accuracy_score(y_test, dt_y_pred_test)

print('the accuracy of the model is:',dt_accuracy)
```

```python
#test the model
print('the predicted value is:',getWeather(dt_model.predict(x_test[3:4])))
print(x_test[3:4])  
```

```python
#show the decision tree
plt.figure(figsize=(100, 50))
tree.plot_tree(dt_model, feature_names=['precipitation', 'temp_max', 'temp_min', 'wind'])
plt.show()
```

# Naive

```python
#make the naive model
nv_model = GaussianNB()
nv_model.fit(x_train_balanced, y_train_balanced)
nv_y_pred_test = nv_model.predict(x_test)
nv_accuracy = accuracy_score(y_test, nv_y_pred_test)

print('the accuracy of the model is:',nv_accuracy)
```

```python
#test the model
print('the predicted value is:',getWeather(nv_model.predict(x_test[3:4])))
print(x_test[3:4])  
```

# Logistic Regression

```python
logReg_model = LogisticRegression(penalty='l1',solver='liblinear',random_state=0)
logReg_model.fit(x_train_balanced, y_train_balanced)

lr_y_pred_test = logReg_model.predict(x_test)
lr_accuracy = accuracy_score(y_test, lr_y_pred_test)

print('the accuracy of the model is:',lr_accuracy)
```

```python
#test the model
print('the predicted value is:',getWeather(logReg_model.predict(x_test[3:4])))
print(x_test[3:4])  
```

# Compare between the three models accuracy 

```python
labels = ['Logistic Regression', 'Naive' ,'Decision Tree']
sizes = [lr_accuracy, nv_accuracy, dt_accuracy ]
explode = (0.05, 0.2, 0.15)
ttt,fig = plt.subplots()
fig.pie(sizes, explode = explode, labels=labels, autopct='%1.1f%%')
plt.show()
```

```python
data = {
    'Models':['Logistic Regression', 'Naive' ,'Decision Tree'],
    'Accuracy': [lr_accuracy, nv_accuracy, dt_accuracy ]
}
df = pd.DataFrame(data)

ax = df.plot.bar(x='Models', y='Accuracy', rot=0)
ax.set_ylim([0, 1])
ax.set_title('Model Accuracy Comparison')
ax.set_xlabel('Models')
ax.set_ylabel('Accuracy')

plt.show()
```
